```python
ratings=pd.read_csv("C:\\Users\\praha\\Desktop\\MTech 1-2 LECS\\LAB\\pavement lab\\BX-CSV-Dump\\BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding="latin-1")
```

```python
x=ratings['User-ID'].value_counts()>200
y=x[x].index
y
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
r2=ratings
r2
```

```python
books_with_rating=ratings.merge(books,on='ISBN')
books_with_rating
```

```python
number_rating=books_with_rating.groupby("Title")['Book-Rating'].count().reset_index()
```

```python
number_rating.rename(columns={'Book-Rating':"number_of_ratings"},inplace=True)
```

```python
final_rating=books_with_rating.merge(number_rating,on="Title")
```

```python
final_rating=final_rating[final_rating["number_of_ratings"]>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(columns="User-ID",index="Title",values="Book-Rating")
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
np.where(book_pivot.index==237)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[184,:].values.reshape(1,-1),n_neighbors=6)
```

```python
def recommend_book(book_name):
    id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestions)):
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book("Animal Farm")
```
